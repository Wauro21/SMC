## Screw regresion

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Load data and visualize data

```python
# Load data
screw_data = np.genfromtxt('screw_data.csv', delimiter=',', skip_header=1)

plt.scatter(screw_data[:,0], screw_data[:,1])
plt.grid()
plt.xlabel('Screw turns [-]')
plt.ylabel('Screw external length [mm]')
plt.title('Relationship betweens turns and external length')
plt.show()
```

## Clean the data and get linear regression

Datapoint (11,59.96) can be considered as an outlier therefore can be removed before calculating the linear regression.

```python
# Clean data
clean_data = np.delete(screw_data, 11, axis=0)

# linear fit 
screw_line = np.polyfit(clean_data[:,0], clean_data[:,1], 1)
line_x = np.linspace(0,20,21)
line_values = np.polyval(screw_line, line_x)

print('Turn per lenght relationship: {} mm/turn'.format(screw_line[0]))
```

## Plot results

```python
# Cleaned data
plt.scatter(clean_data[:,0], clean_data[:,1], label= 'Cleaned data')

# Line plot
plt.plot(line_x, line_values, color='red', label='Linear fit')

plt.grid()
plt.xlabel('Screw turns [-]')
plt.ylabel('Screw external length [mm]')
plt.title('Relationship betweens turns and external length')
plt.legend()
plt.show()
```
